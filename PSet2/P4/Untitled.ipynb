```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
canvas = np.zeros(shape=(500, 200), dtype=int)
canvas_height = np.zeros(shape=(200,), dtype=int, order='F')

color_map = 1
count = 0

canvas.shape
```

```python
for _ in range(50000):
    if count == 2000:
        color_map = - color_map
        count = 0
    rand = np.random.randint(0, 200)
    canvas_height[rand] += 1
    count += 1
    canvas[canvas_height[rand]][rand] = color_map
```

```python
fig, ax = plt.subplots(1, 1)
ax.pcolor(canvas, cmap='RdBu')
ax.set_ylim(0, np.max(canvas_height) + 10)
plt.show()
```
