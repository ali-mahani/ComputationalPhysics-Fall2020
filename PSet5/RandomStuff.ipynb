```python
import numpy as np
from matplotlib import pyplot as plt
```

# Problem 1

```python
N = 10000
my_rand = np.random.randint(10, size=N)
counts, bins = np.histogram(my_rand)
# graphics
plt.hist(my_rand, bins=10)
plt.savefig("p1_hist.jpg", bbox_inches='tight')
plt.show()
plt.close()
```

```python
y_data = []
for N in range(500, 1501, 100):
    rands = np.random.randint(10, size=N)
    y_data.append(np.sqrt(np.var(rands)) / float(N))

x_data = np.linspace(500, 1500, 11)
y_data = np.array(y_data)
y_data.shape

plt.plot(np.sqrt(x_data ** -1), y_data)
plt.xlabel('1 / sqrt(N)')
plt.ylabel('stdev / N')
plt.savefig("p1_plot.jpg", bbox_inches='tight')
plt.close()
```

# Problem 3

```python
N = np.array([5, 10, 100, 1000])

hist_data = []
for n in N:
    data = np.zeros((1000, ), dtype=int)
    for i in range(1000):
        data[i] = np.sum(np.random.randint(10, size=n))
    hist_data.append(data)

for i in range(2):
    plt.hist(hist_data[i], bins=15)
    plt.title('for N = ' + str(N[i]))
    plt.savefig('p3_' + str(N[i]) + '.jpg', bbox_inches='tight')
    plt.show()

for i in range(2, 4):
    plt.hist(hist_data[i], bins=30)
    plt.title('for N = ' + str(N[i]))
    plt.savefig('p3_' + str(N[i]) + '.jpg', bbox_inches='tight')
    plt.show()
```

# Problem 4

```python

for N in [10, 100, 500, 1000]:
    sigma = 1
    theta = np.random.uniform(-np.pi, np.pi, N)
    x = np.sqrt(-2 * sigma ** 2 * np.log(np.random.uniform(0, 1, size=N))) * np.cos(theta)
    y = np.sqrt(-2 * sigma ** 2 * np.log(np.random.uniform(0, 1, size=N))) * np.sin(theta)
    plt.scatter(x, y)
    plt.title('2D Normal Random generation ' + str(N))
    plt.savefig('p4_' + str(N) + '.jpg', bbox_inches='tight')
    plt.show()
    plt.close()



```
